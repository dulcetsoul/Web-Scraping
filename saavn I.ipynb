```python
from bs4 import BeautifulSoup
import requests

res = requests.get("https://www.jiosaavn.com/featured/weekly-top-songs");
soup = BeautifulSoup(res.text,'lxml');
data = soup.find('ol',{'class':'content-list'})
all_songs = data.find_all('div',{'class':'details'})
all_songs
```

```python
for count,s in enumerate(all_songs,1):
    song = s.find("p",{"class":"song-name"})
    album = s.find("p",{"class":"album-name"})
    print(str(count) + " Song Name: " + song.text + "  Album Name: " + album.text)
    
```
